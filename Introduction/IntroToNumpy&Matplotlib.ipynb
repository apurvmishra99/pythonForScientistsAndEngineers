```python
x = np.arange(0.0, 20.0, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, '-g', label = 'sine')
plt.plot(x, y2, '-r', label = 'cos')
plt.grid(b=True)
plt.legend(loc = 'upper right')
plt.show()
```
